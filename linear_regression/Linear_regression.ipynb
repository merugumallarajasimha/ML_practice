# MULTIPLE LINEAR REGRESSION

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv(r"C:\Users\merug\Desktop\StudentPerformance.csv")
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(" ","_")
df.columns=df.columns.str.title()
```

```python
df.columns
```

```python
df['Performance_Index']=df['Performance_Index'].astype(int)
# converted this col to int because all values we in int datatype , no decimals were present
```

```python
df['Extracurricular_Activities'] = df['Extracurricular_Activities'].map({'Yes': 1, 'No': 0}) # instead of OneHotEncoder I coverted it in binary
```

```python
x=df[['Hours_Studied','Previous_Scores','Extracurricular_Activities','Sleep_Hours','Sample_Question_Papers_Practiced']] #Independent_Variable # INPUT
y=df['Performance_Index']# Dependemt_Variable #OUTPUT
```

```python
# Our prediction of  Prefromace_Index is completey based on the all independent cols
```

```python
lr=LinearRegression()
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2)
```

```python
lr.fit(X_train,Y_train)
```

```python
lr.predict(X_test)
```

```python
lr.score(X_test,Y_test)
```
